# `Import the main Libraries`

```python
## Major Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import warnings

warnings.filterwarnings("ignore")
%load_ext nb_black
```

```python
# Path dataset
path_versoin_1 = r"D:\Software_Courses\Coding\Data Science Project\life_expectancy\data\processed/data_version_1.pkl"
```

```python
# Read data and print head
df = pd.read_pickle(path_versoin_1)
df.head()
```

# `Univariate Analysis`

```python
# Get summery statistics for the numerical column
df.describe().T.style.background_gradient(cmap="Blues")
```

```python
# distplot Distribution of Numerical(continues) Variables in Dataset
col_discrete = ["schooling", "year"]
plt.figure(figsize=(10, 20))
count = 0
for num, name in enumerate(df.select_dtypes(exclude="object")):
    if name in (col_discrete):
        continue
    else:
        plt.subplot(7, 3, count + 1)
        sns.distplot(df[name])
        plt.title(name)
        plt.tight_layout()
        count += 1
plt.show()
```

```python
## Countplot of categorical Variables in Dataset
# Set the figure size
plt.figure(figsize=(20, 200))

# Create the count plot
for i, name in enumerate(["status", "schooling"]):
    plt.subplot(24, 2, i + 1)
    sns.countplot(x=df[name], order=df[name].value_counts().index, palette="rocket")

    # Set labels and title
    plt.title(f"Countplot Distribution of {name}", color="red", fontsize=25)
    plt.xlabel(name, color="red", fontsize=25)
    plt.ylabel("Count", color="red", fontsize=25)

# Display the plot
plt.show()
```

# `1. How does life expectancy vary across different countries and years?`

```python
# How does life expectancy vary across different years?
# Set the figure size
plt.figure(figsize=(8, 4))

# Create the line plot
sns.lineplot(data=df, x="year", y="life_expectancy")

# Set labels and title
plt.xlabel("Year", color="red")
plt.ylabel("Life Expectancy", color="red")
plt.title("Life Expectancy Variation Across  Years", color="red")

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# What is the country that has the highest average life expectancy?
df[df.life_expectancy == df.life_expectancy.max()]
```

```python
##  How does life expectancy vary across different the country that has the highest average life expectancy and years?
# Set the figure size
plt.figure(figsize=(13, 6))

# Create the line plot
sns.lineplot(data=df, x="year", y="life_expectancy",hue=df[df['country'].isin(df[df.life_expectancy == df.life_expectancy.max()]['country'].tolist())]['country'])

# Set labels and title
plt.xlabel("Year", color="red")
plt.ylabel("Life Expectancy", color="red")
plt.title("Life Expectancy Variation Across the country that has the highest average life expectancy and Years", color="red")

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# Which country has the lowest life expectancy?
df[df.life_expectancy == df.life_expectancy.min()]
```

```python
##  How does life expectancy vary across different the country that has the lowest average life expectancy and years?
# Set the figure size
plt.figure(figsize=(8, 4))

# Create the line plot
sns.lineplot(data=df, x="year", y="life_expectancy",hue=df[df['country'].isin(df[df.life_expectancy == df.life_expectancy.min()]['country'].tolist())]['country'])

# Set labels and title
plt.xlabel("Year", color="red")
plt.ylabel("Life Expectancy", color="red")
plt.title("Life Expectancy Variation Across the country that has the lowest average life expectancy and Years", color="red")

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

# `5. Is there a relationship between healthcare expenditure(percentage_expenditure && total_expenditure ) and life expectancy?`

```python
# Calculate the correlation coefficient between life expectancy and percentage_expenditure
correlation = df["percentage_expenditure"].corr(df["life_expectancy"])    ## print correlation on the Visualization

# Set the figure size
plt.figure(figsize=(6, 3))

# Visualize the relationship using a scatter plot
sns.scatterplot(data=df, x="percentage_expenditure", y="life_expectancy")

# Set labels and title
plt.title(f"Correlation is ({correlation:.2f})", color="g")
plt.xlabel("percentage_expenditure", color="red")
plt.ylabel("life_expectancy", color="red")

# Display the plot
plt.show()
```

```python
# Calculate the correlation coefficient between life expectancy and ( total_expenditure )
correlation = df["total_expenditure"].corr(df["life_expectancy"])    ## print correlation on the Visualization

# Set the figure size
plt.figure(figsize=(6, 3))

# Visualize the relationship using a scatter plot
sns.scatterplot(data=df, x="total_expenditure", y="life_expectancy")

# Set labels and title
plt.title(f"Correlation is ({correlation:.2f})", color="g")
plt.xlabel("total_expenditure", color="red")
plt.ylabel("life_expectancy", color="red")

## Grid
sns.set_theme(style="whitegrid")

# Display the plot
plt.show()
```

**`5-1`Is there a relationship between healthcare expenditure(c && total_expenditure ) and life expectancy?`in Countries with the highest life expectancy`**

```python
# Calculate the Countries with the highest life expectancy
highest_countries = df[
    df["life_expectancy"] == df["life_expectancy"].max()
].country.values.tolist()

# New dataset include the Countries with the highest life expectancy

df_highest = df[df["country"].isin(highest_countries)]

# Calculate the correlation coefficient between life expectancy and total_expenditure
correlation = df_highest["total_expenditure"].corr(df_highest["life_expectancy"])  ## print correlation on the Visualization


# Visualize the relationship using a line plot
sns.lineplot(data=df_highest, x="total_expenditure", y="life_expectancy", hue="country" )

# Set labels and title
plt.title(f"Correlation is ({correlation:.2f})", color="g")
plt.xlabel("total_expenditure", color="red")
plt.ylabel("life_expectancy", color="red")

# place legend in upper left of plot
plt.legend(loc="lower right", fontsize=8)

## Grid
sns.set_theme(style="whitegrid")

# Display the plot
plt.show()
```

```python
# Calculate the correlation coefficient between life expectancy and  percentage_expenditure
correlation = df_highest["percentage_expenditure"].corr(df_highest["life_expectancy"])    ## print correlation on the Visualization


# Visualize the relationship using a line plot
sns.lineplot(data=df_highest, x="percentage_expenditure", y="life_expectancy",hue='country')

# Set labels and title
plt.title(f"Correlation is ({correlation:.2f})", color="g")
plt.xlabel("percentage_expenditure", color="red")
plt.ylabel("life_expectancy", color="red")

#place legend in upper left of plot
plt.legend(loc='upper left',fontsize=8)

# Display the plot
plt.show()
```
